```python
dic = {}
with open(all_celltype_filepath,'r',encoding="ISO-8859-1") as f:
    content = f.readlines()
    for line in content:
        line_list = line.strip().split('\t')
        dic[int(line_list[0][1:])] = '_'.join(line_list[1:])
#         print(line.strip().split('\t'))
```

# select blood-cell type from overlapMatrix

```python
test_overlapMatrix_filepath = '/home/shuang/projects/development_eqtm/data/testData/2017-12-09-eQTLsFDR-et0.0-flipped_overlapMatrix.txt'
test_overlapMatrix = pd.read_csv(test_overlapMatrix_filepath,index_col=0)
```

```python
# blood_celltype
blood_features = []
for featurename in test_overlapMatrix.columns:
    for celltype in blood_celltype:
        if celltype in featurename:
            blood_features.append(featurename)
test_overlapMatrix_bloodCelltype = test_overlapMatrix[blood_features].copy()
test_overlapMatrix_nonBlood = test_overlapMatrix[[col for col in test_overlapMatrix.columns if col not in blood_features]].copy()

featurelist_filepath = '/home/shuang/projects/development_eqtm/data/testData/2017-12-09-eQTLsFDR-et0.0-flipped_featureList.txt'
featurelist = open(featurelist_filepath).readlines()
print(len(featurelist),featurelist[-1])

histonelist_filepath = '/home/shuang/projects/development_eqtm/data/testData/2017-12-09-eQTLsFDR-et0.0-flipped_histoneList.txt'
histonelist = [col.strip() for col in open(histonelist_filepath,'r').readlines()]
print(len(histonelist),histonelist[0])

overlapRatio_blood = pd.DataFrame(data=0,
                                  index=test_overlapMatrix.index,
                                  columns = histonelist)
overlapRatio_nonBlood = pd.DataFrame(data=0,
                                  index=test_overlapMatrix.index,
                                  columns = histonelist)
num_bloodCellType = len(blood_celltype)
num_nonBlood = len(celltype_list_fromfeaturedir)-num_bloodCellType
print(histonelist)
for col in test_overlapMatrix_bloodCelltype.columns:
    histone_name = col.split('-')[1]
    overlapRatio_blood[histone_name] += test_overlapMatrix[col]/num_bloodCellType
for col in test_overlapMatrix_nonBlood.columns:
    histone_name = col.split('-')[1]
    overlapRatio_nonBlood[histone_name] += test_overlapMatrix_nonBlood[col]/num_nonBlood
print(overlapRatio_blood.head())
print(overlapRatio_nonBlood.head())
overlapRatio_blood.to_csv('/home/shuang/projects/development_eqtm/data/testData/2017-12-09-eQTLsFDR-et0.0-flipped_blood.txt')
overlapRatio_nonBlood.to_csv('/home/shuang/projects/development_eqtm/data/testData/2017-12-09-eQTLsFDR-et0.0-flipped_nonBlood.txt')
```

```python
test_overlapMatrix_filepath = '/home/shuang/projects/development_eqtm/data/eqtmZscores/overlapMatrix/2017-12-09-eQTLsFDR-et0.0-flipped_overlapMatrix.txt'
test_overlapMatrix = pd.read_csv(test_overlapMatrix_filepath,index_col=0)

celltype_list_filepath = '/home/shuang/projects/development_eqtm/data/testData/cell_list.txt'
celltype_list_fromfeaturedir = list(filter(None,
                                           [int(row.strip()[1:]) for row in open(celltype_list_filepath,'r').readlines()]))

featurelist_filepath = '/home/shuang/projects/development_eqtm/data/testData/2017-12-09-eQTLsFDR-et0.0-flipped_featureList.txt'
featurelist = open(featurelist_filepath).readlines()
print(len(featurelist),featurelist[-1])

histonelist_filepath = '/home/shuang/projects/development_eqtm/data/testData/2017-12-09-eQTLsFDR-et0.0-flipped_histoneList.txt'
histonelist = [col.strip() for col in open(histonelist_filepath,'r').readlines()]
print(len(histonelist),histonelist[0])

overlapRatio_all = pd.DataFrame(data=0,
                                  index=test_overlapMatrix.index,
                                  columns = histonelist)

num_allCelltypes = len(celltype_list_fromfeaturedir)
for col in test_overlapMatrix.columns:
    histone_name = col.split('-')[1]
    overlapRatio_all[histone_name] += test_overlapMatrix[col]/num_allCelltypes
print(overlapRatio_all.head())
overlapRatio_all.to_csv('/home/shuang/projects/development_eqtm/data/eqtmZscores/overlapMatrix/2017-12-09-eQTLsFDR-et0.0-flipped_allCelltypes.txt')
```
